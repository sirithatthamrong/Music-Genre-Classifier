```python
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
# read files
data_3sec = pd.read_csv("features_3_sec.csv")
data_30sec = pd.read_csv("features_30_sec.csv")
```

```python
# All columns
columns_of_interest = ['chroma_stft_mean', 'chroma_stft_var', 
                       'rms_mean', 'rms_var',
                       'spectral_centroid_mean', 'spectral_centroid_var',
                       'spectral_bandwidth_mean',	'spectral_bandwidth_var',
                       'rolloff_mean', 'rolloff_var',
                       'zero_crossing_rate_mean',	'zero_crossing_rate_var',
                       'harmony_mean', 'harmony_var',
                       'perceptr_mean', 'perceptr_var',
                       'tempo', 
                       'mfcc1_mean', 'mfcc1_var',
                       'mfcc2_mean', 'mfcc2_var',
                       'mfcc3_mean', 'mfcc3_var',
                       'mfcc4_mean', 'mfcc4_var',
                       'mfcc5_mean', 'mfcc5_var', 
                       'mfcc6_mean', 'mfcc6_var',
                       'mfcc7_mean', 'mfcc7_var', 
                       'mfcc8_mean', 'mfcc8_var',
                       'mfcc9_mean', 'mfcc9_var', 
                       'mfcc10_mean', 'mfcc10_var',
                       'mfcc11_mean', 'mfcc11_var', 
                       'mfcc12_mean', 'mfcc12_var',
                       'mfcc13_mean', 'mfcc13_var', 
                       'mfcc14_mean', 'mfcc14_var',
                       'mfcc15_mean', 'mfcc15_var', 
                       'mfcc16_mean', 'mfcc16_var',
                       'mfcc17_mean', 'mfcc17_var', 
                       'mfcc18_mean', 'mfcc18_var',
                       'mfcc19_mean', 'mfcc19_var', 
                       'mfcc20_mean', 'mfcc20_var'
                       ]  

# Data of each genre, 3sec and 30sec 
Blue_3s = data_3sec[data_3sec['label'] == 'blues']
Blue_30s = data_30sec[data_30sec['label'] == 'blues']
Classical_3s = data_3sec[data_3sec['label'] == 'classical']
Classical_30s = data_30sec[data_30sec['label'] == 'classical']
Country_3s = data_3sec[data_3sec['label'] == 'country']
Country_30s = data_30sec[data_30sec['label'] == 'country']
Disco_3s = data_3sec[data_3sec['label'] == 'disco']
Disco_30s = data_30sec[data_30sec['label'] == 'disco']
Hiphop_3s = data_3sec[data_3sec['label'] == 'hiphop']
Hiphop_30s = data_30sec[data_30sec['label'] == 'hiphop']
Jazz_3s = data_3sec[data_3sec['label'] == 'jazz']
Jazz_30s = data_30sec[data_30sec['label'] == 'jazz']
Metal_3s = data_3sec[data_3sec['label'] == 'metal']
Metal_30s = data_30sec[data_30sec['label'] == 'metal']
Pop_3s = data_3sec[data_3sec['label'] == 'pop']
Pop_30s = data_30sec[data_30sec['label'] == 'pop']
Reggae_3s = data_3sec[data_3sec['label'] == 'reggae']
Reggae_30s = data_30sec[data_30sec['label'] == 'reggae']
Rock_3s = data_3sec[data_3sec['label'] == 'rock']
Rock_30s = data_30sec[data_30sec['label'] == 'rock']
```

```python
def dataLists(collected_data):
    toRet = {}
    for column in columns_of_interest:
      toRet[column] = collected_data[column].tolist()
    return toRet

# In order to access a specific category -> 
# chroma_stft_mean_data = Blue_3s_data['chroma_stft_mean']
Blue_3s_data = dataLists(Blue_3s[columns_of_interest])
Blue_30s_data = dataLists(Blue_30s[columns_of_interest])
Classical_3s_data = dataLists(Classical_3s[columns_of_interest])
Classical_30s_data = dataLists(Classical_30s[columns_of_interest])
Country_3s_data = dataLists(Country_3s[columns_of_interest])
Country_30s_data = dataLists(Country_30s[columns_of_interest])
Disco_3s_data = dataLists(Disco_3s[columns_of_interest])
Disco_30s_data = dataLists(Disco_30s[columns_of_interest])
Hiphop_3s_data = dataLists(Hiphop_3s[columns_of_interest])
Hiphop_30s_data = dataLists(Hiphop_30s[columns_of_interest])
Jazz_3s_data = dataLists(Jazz_3s[columns_of_interest])
Jazz_30s_data = dataLists(Jazz_30s[columns_of_interest])
Metal_3s_data = dataLists(Metal_3s[columns_of_interest])
Metal_30s_data = dataLists(Metal_30s[columns_of_interest])
Pop_3s_data = dataLists(Pop_3s[columns_of_interest])
Pop_30s_data = dataLists(Pop_30s[columns_of_interest])
Reggae_3s_data = dataLists(Reggae_3s[columns_of_interest])
Reggae_30s_data = dataLists(Reggae_30s[columns_of_interest])
Rock_3s_data = dataLists(Rock_3s[columns_of_interest])
Rock_30s_data = dataLists(Rock_30s[columns_of_interest])
```

```python
plt.figure(figsize=(14, 7))
plt.plot(Blue_3s_data['chroma_stft_mean'], Blue_3s_data['chroma_stft_var'], '.', color='orange', label='blue', alpha=0.3)
plt.plot(Classical_3s_data['chroma_stft_mean'], Classical_3s_data['chroma_stft_var'], '.', color='green', label='classical', alpha=0.3)
plt.plot(Country_3s_data['chroma_stft_mean'], Country_3s_data['chroma_stft_var'], '.', color='blue', label='country', alpha=0.3)
plt.plot(Disco_3s_data['chroma_stft_mean'], Disco_3s_data['chroma_stft_var'], '.', color='red', label='disco', alpha=0.3)
plt.plot(Hiphop_3s_data['chroma_stft_mean'], Hiphop_3s_data['chroma_stft_var'], '.', color='purple', label='hiphop', alpha=0.3)
plt.plot(Jazz_3s_data['chroma_stft_mean'], Jazz_3s_data['chroma_stft_var'], '.', color='magenta', label='Jazz', alpha=0.3)
plt.plot(Metal_3s_data['chroma_stft_mean'], Metal_3s_data['chroma_stft_var'], '.', color='yellow', label='metal', alpha=0.3)
plt.plot(Pop_3s_data['chroma_stft_mean'], Pop_3s_data['chroma_stft_var'], '.', color='gray', label='pop', alpha=0.3)
plt.plot(Reggae_3s_data['chroma_stft_mean'], Reggae_3s_data['chroma_stft_var'], '.', color='brown', label='reggae', alpha=0.3)
plt.plot(Rock_3s_data['chroma_stft_mean'], Rock_3s_data['chroma_stft_var'], '.', color='cyan', label='rock', alpha=0.3)

plt.xlabel('Chroma STFT Mean')
plt.ylabel('Chroma STFT Variance')
plt.title('Scatter Plot of Chroma STFT Mean vs. Variance')
plt.legend()
plt.grid()
```
